Q1. Discuss string slicing and provide examples.

String slicing is a technique in Python used to extract a specific part or subset of a string. Strings are sequences of characters, and slicing allows you to select a portion of that sequence by specifying a start and an end index.

Basic syntax
string[start:end:step]


 Examples


```python
string = "hello world"
string[0:5]   #by default step is one
```

```python
string[0:5:2]
```

```python
string[-5:]  #slices the last five words
```

```python
string[::-1]  #slices the entire string in reverse
```

Q4. Compare and contrast tuples and lists with examples.

Mutability

Lists: Mutable, meaning their elements can be changed, added, or removed after creation.
  
Tuples: Immutable, meaning once they are created, their elements cannot be changed, added, or removed.


```python
l = [1, 2, 3]
t = (1, 2, 3)
l[0] = 4
l
```

```python
t[0] = 4
t
```

Use Cases

Lists: Used when you need a collection that can be modified, such as when items need to be added, removed, or updated.

Tuples: Used when you need a fixed collection of items, such as constants, or when you want to ensure the data cannot be changed.



 Methods and Operations

Lists: Provide a variety of methods for modification, such as append(), remove(), pop(), sort(), etc.

Tuples: Provide fewer methods, mainly focusing on operations like counting occurrences (count()) and finding indices (index()).


```python
l.append(5)
l
```

```python
l.sort()
l
```

```python
t = (1, 2, 2, 3, 3, 3, 3)
t.count(3)
```

```python
t.index(2)
```

 Immutability Benefits

Lists: Mutability can lead to unintended side effects if the list is altered in different parts of a program.

Tuples: Immutability makes tuples hashable, meaning they can be used as keys in dictionaries or elements in sets.


Q5. Describe the features of sets and provide exsamples of their use.


Sets in Python are a built-in data structure that represent an unordered collection of unique elements. They are highly useful for operations involving membership testing, removing duplicates, and performing mathematical set operations like union, intersection, and difference.

Key Features of Sets
1. Unordered Collection

 Sets are unordered, meaning the elements do not have a defined sequence. When you iterate over a set or print it, the elements may appear in any order.



```python
set = {2, 4, 3, 1}
set
```

2. Unique Elements

 Sets automatically ensure that all elements are unique. If you try to add a duplicate element, it will be ignored.

```python
s = {1, 2, 3, 4, 4, 5, 3, 5, 1}
s
```

3. Mutable

 Sets are mutable, allowing you to add, remove, or modify elements after the set has been created. However, the elements themselves must be immutable (e.g., numbers, strings, tuples).

```python
set = {1, 2, 3}
set.add(4)    # Adding an element
set.remove(2) # Removing an element
set
```

4. No Indexing or Slicing

 Unlike lists or tuples, sets do not support indexing, slicing, or other sequence-like behavior because they are unordered.

```python
set[0]
```

5. Set Operations

Sets support mathematical operations such as union, intersection, difference, and symmetric difference.

```python
#union(|): Combines elements from both sets, removing duplicates.
set1 = {1, 2, 3}
set2 = {3, 4, 5}
union_set = set1 | set2
print(union_set)
```

```python
#Intersection (&): Returns only the elements common to both sets.
intersection_set = set1 & set2
print(intersection_set)
```

```python
#Difference (-): Returns elements that are in the first set but not in the second.
set1 = {1, 2, 3}
set2 = {2, 3, 4}
difference_set = set1 - set2
print(difference_set)
```

```python
#Symmetric Difference (^): Returns elements that are in either of the sets, but not in both.
sym_diff_set = set1 ^ set2
print(sym_diff_set)
```

6. Efficient Membership Testing

Sets are optimized for checking whether an element is present in the set. Membership tests are faster with sets compared to lists or tuples.

```python
s = {1, 2, 3, 4, 5}
print(1 in s)
print(6 in s)
```

7. Set Comprehensions

Similar to list comprehensions, you can use set comprehensions to create a new set based on an existing iterable.

```python
Squares = {x**2 for x in range(5)}
Squares
```

Q6 Discuss the use cases of tuples and sets in Python Programming

Use Cases of Tuples
1. Storing Fixed Data:
Tuples are ideal for storing collections of values that should remain constant throughout the program. For example, they can be used to store coordinates or configuration settings that are not meant to change.

2. Returning Multiple Values from Functions:
Tuples are commonly used to return multiple values from a function. This allows a function to return a group of related data points in a single, immutable collection.

3. Immutable Data Structures:
Since tuples are immutable, they are suitable for scenarios where data integrity is important. For example, they can be used as keys in dictionaries because their values cannot be changed.

4. Memory Efficiency:
Tuples are more memory-efficient than lists, making them suitable for storing large amounts of read-only data.

5. Passing Immutable Arguments to Functions:
When passing data to functions, tuples ensure that the data cannot be altered within the function, preserving the original values.

Use Cases of Sets
1. Removing Duplicates:
Sets are perfect for filtering out duplicate elements from a collection. When you need a unique list of items, converting a list to a set automatically removes duplicates.

2. Efficient Membership Testing:
Sets are optimized for checking whether an element is present in the set, making them efficient for tasks like checking membership in a collection of values.

3. Performing Set Operations:
Sets are ideal for performing mathematical operations like union, intersection, and difference, which are useful in scenarios like comparing datasets or filtering data.

4. Filtering Data:
Sets can be used to filter collections by intersecting them with a set of allowed values. This is useful when you need to ensure that only certain elements are retained in a collection.

5. Handling Unordered Collections:
When the order of elements is not important, and you need to ensure all elements are unique, sets are the right choice. They are especially useful for managing collections like tags or categories in content management systems.

Q8 Discuss the importance of dictionary keys being immutable and provide examples.


In Python, dictionary keys must be immutable, meaning their value cannot change after they are created. This immutability is crucial for several reasons, primarily because it ensures that dictionary keys are hashable and that the integrity of the data structure is maintained.

 Hashability

 Dictionary keys are stored using a hashing mechanism, which allows Python to quickly locate values associated with keys. For a key to be hashable, it must have a consistent hash value during its lifetime. If a key were mutable, its hash value could change, leading to difficulties in retrieving the associated value.


```python
dict = {"name":"XYZ"} #immutable string
dict['name']
```

```python
dict = {(1,2,3):"XYZ"} #immutable tuple
dict[(1,2,3)]
```

```python
dict = {[1,2,3]:"XYZ"} #mutable list
dict
```

Ensuring Uniqueness

Dictionary keys must be unique, and immutability helps enforce this rule. If keys were mutable, they could change in such a way that two keys could become identical, violating the uniqueness constraint and leading to undefined behavior.

Efficiency in Lookup Operations

Dictionaries rely on efficient lookup operations to retrieve values based on keys. If keys were mutable, each lookup could potentially require rehashing or checking the entire dictionary, which would negate the efficiency benefits of the hash table.

Example:
Using an immutable key like an integer or string ensures that dictionary lookups remain fast and efficient, as Python can directly calculate the hash and find the corresponding value.

Examples of Immutable and Mutable Types

Immutable Types:
Strings, numbers (int, float), tuples, and frozensets are common examples of immutable types that can be used as dictionary keys.

Mutable Types:
 Lists, sets, and dictionaries themselves are mutable and cannot be used as dictionary keys.
